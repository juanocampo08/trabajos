```python
import random

class Planta:
    def __init__(self):
        self.energia = 5

    def __str__(self):
        return "M"

class Presa:
    def __init__(self, x, y):
        self.energia = 5
        self.x = x
        self.y = y
        self.viva = True

    def moverse(self, matriz):
      direccion = buscar_planta(self, matriz, 1)

      if direccion:
        mover_presa_recursivo(self, matriz, [direccion], 0)
      else:
        mover_presa_recursivo(self, matriz, [(0, 1), (0, -1), (1, 0), (-1, 0)], 0)

    def __str__(self):
        return "P"

class Depredador:
    def __init__(self, x, y):
        self.energia = 10
        self.x = x
        self.y = y
        self.vivo = True

    def moverse(self, matriz):
        direccion = buscar_presa(self, matriz, 1)
        if direccion:
            mover_depredador_recursivo(self, matriz, [direccion], 0)
        else:
            mover_depredador_recursivo(self, matriz, [(0, 1), (0, -1), (1, 0), (-1, 0)], 0)

    def morirse(self, matriz):
      muerte_depredador_recursivo(self, matriz, True)

    def __str__(self):
        return "D"

def buscar_planta(presa, matriz, alcance, direccion_index = 0):
  if alcance > len(matriz):
    return None

  direcciones = [(0, 1), (0, -1), (1, 0), (-1, 0)]

  if direccion_index >= len(direcciones):
    return buscar_planta(presa, matriz, alcance + 1, 0)

  dx, dy = direcciones[direccion_index]
  nueva_x = presa.x + dx * alcance
  nueva_y = presa.y + dy * alcance

  if 0 <= nueva_x < len(matriz) and 0 <= nueva_y < len(matriz):
    if isinstance(matriz[nueva_x][nueva_y], Planta):
      return dx, dy

  return buscar_planta(presa, matriz, alcance, direccion_index + 1)


def buscar_presa(depredador, matriz, alcance, direccion_index=0):
    if alcance > len(matriz):
        return None

    direcciones = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    if direccion_index >= len(direcciones):
        return buscar_presa(depredador, matriz, alcance + 1, 0)

    dx, dy = direcciones[direccion_index]
    nueva_x = depredador.x + dx * alcance
    nueva_y = depredador.y + dy * alcance

    if 0 <= nueva_x < len(matriz) and 0 <= nueva_y < len(matriz):
        if isinstance(matriz[nueva_x][nueva_y], Presa):
            return dx, dy

    return buscar_presa(depredador, matriz, alcance, direccion_index + 1)

def generar_matriz(n, i=0, j=0, fila=None, matriz=None):
    if fila is None:
        fila = []
    if matriz is None:
        matriz = []
    if i == n:
        return matriz
    if j == n:
        matriz.append(fila)
        return generar_matriz(n, i + 1, 0, [], matriz)

    rand = random.random()
    if rand < 0.1:
        fila.append(Depredador(i, j))
    elif rand < 0.5:
        fila.append(Presa(i, j))
    elif rand < 0.8:
        fila.append(Planta())
    else:
        fila.append(None)

    return generar_matriz(n, i, j + 1, fila, matriz)

def contar_plantas(matriz, i = 0, j = 0, contador =  0):

  if i == len(matriz):
    return contador

  if j == len(matriz[i]):
    return contar_plantas(matriz, i + 1, 0, contador)

  if isinstance(matriz[i][j], Planta):
    contador += 1

  return contar_plantas(matriz, i, j + 1, contador)

def agregar_plantas(matriz, i = 0, j = 0, minimo = 3, plantas_agregadas = 0):
  plantas_actuales = contar_plantas(matriz)
  if plantas_actuales >= minimo or i >= len(matriz):
    return #no es necesario agregar plantas

  if j == len(matriz[i]):
    return agregar_plantas(matriz, i + 1, j, minimo, plantas_agregadas)

  if matriz[i][j] is None:
    matriz [i][j] = Planta()
    plantas_agregadas += 1

  return agregar_plantas(matriz, i, j + 1, plantas_agregadas)

def imprimir_matriz_recursivo(matriz, i=0, j=0):
    if i >= len(matriz):
        return

    if j >= len(matriz[i]):
        print()
        imprimir_matriz_recursivo(matriz, i + 1, 0)
        return

    if matriz[i][j] is None:
        print(".", end=" ")
    else:
        print(matriz[i][j], end=" ")
    imprimir_matriz_recursivo(matriz, i, j + 1)

def mover_presa_recursivo(presa, matriz, direcciones, index):

    if index == len(direcciones):
        return

    dx, dy = direcciones[index]
    nueva_x = presa.x + dx
    nueva_y = presa.y + dy

    if 0 <= nueva_x < len(matriz) and 0 <= nueva_y < len(matriz):
        if isinstance(matriz[nueva_x][nueva_y], Planta):
            presa.energia += 3
            matriz[nueva_x][nueva_y] = presa
            matriz[presa.x][presa.y] = None
            presa.x = nueva_x
            presa.y = nueva_y
            return

        elif matriz[nueva_x][nueva_y] is None:
            matriz[nueva_x][nueva_y] = presa
            matriz[presa.x][presa.y] = None
            presa.x = nueva_x
            presa.y = nueva_y
            return

    mover_presa_recursivo(presa, matriz, direcciones, index + 1)

def mover_depredador_recursivo(depredador, matriz, direcciones, index):
    if index == len(direcciones):
        return

    dx, dy = direcciones[index]

    nueva_x = depredador.x + dx
    nueva_y = depredador.y + dy

    if 0 <= nueva_x < len(matriz) and 0 <= nueva_y < len(matriz):
        if isinstance(matriz[nueva_x][nueva_y], Presa):
            depredador.energia += 5
            matriz[nueva_x][nueva_y] = depredador
            matriz[depredador.x][depredador.y] = None
            depredador.x = nueva_x
            depredador.y = nueva_y
            return

        elif matriz[nueva_x][nueva_y] is None:
            depredador.energia -= 1
            matriz[nueva_x][nueva_y] = depredador
            matriz[depredador.x][depredador.y] = None
            depredador.x = nueva_x
            depredador.y = nueva_y
            return

    mover_depredador_recursivo(depredador, matriz, direcciones, index + 1)


def muerte_depredador_recursivo(depredador, matriz, vivo):
  if vivo == False:
    return

  if depredador.energia <= 0:
    matriz[depredador.x][depredador.y] = None
    depredador.vivo = False

  muerte_depredador_recursivo(depredador, matriz, False)

def iniciar_programa(matriz, ciclo = 0, ciclos_totales = 10, i = 0, j = 0):
  if ciclo == ciclos_totales:
    return

  if ciclo == 5:
    agregar_plantas(matriz)

  if i == len(matriz):
    print(f"\n Ciclo {ciclo + 1}.")
    imprimir_matriz_recursivo(matriz)
    input()
    return iniciar_programa(matriz, ciclo + 1, ciclos_totales, 0, 0)

  if j == len(matriz[i]):
    return iniciar_programa(matriz, ciclo, ciclos_totales, i + 1, 0)

  entidad = matriz[i][j]
  if isinstance(entidad, Depredador):
    entidad.moverse(matriz)
    entidad.morirse(matriz)
  elif isinstance(entidad, Presa):
    entidad.moverse(matriz)

  return iniciar_programa(matriz, ciclo, ciclos_totales, i, j + 1)

if __name__ == "__main__":
  matriz = generar_matriz(8)
  iniciar_programa(matriz)
```
